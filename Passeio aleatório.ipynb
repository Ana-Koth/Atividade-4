1.Passeio aleatório (implementação básica). Considere um passeio aleatório unidimensional, onde a cada instante de tempo o caminhante dá um passo uma unidade para a direita ou para a esquerda com probabilidades p e 1−p, respectivamente. O passeio termina depois de N=1000 passos. Faça um gráfico da posição do caminhante em função do número de passos (que representa o tempo). Use p=0. e p=0.6. Implemente na forma de uma função que receba como argumento o número de passos N e a probabilidade p, e retorne um array com a posição do caminhante em cada passo

```python
import numpy as np
import matplotlib.pyplot as plt

def Passeio_aleatorio(N):
    """
    Simula um passeio aleatório unidimensional.
    
    Parâmetros:
    n_passos (int): Número de passos no passeio aleatório.
    
    Retorna:
    np.ndarray: Posições ao longo do passeio aleatório.
    """
    # Gera uma sequência de passos aleatórios
    # Cada passo pode ser -1 ou 1 com igual probabilidade
    # Gera gráfico do passeio aleatório
    
    N = 1000 # Número de passos
    passos1 = np.random.choice([-1, 1], size=N, p=[0.6 , 0.4]) # Gera uma sequência de passos aleatórios
    posições1 = np.cumsum(passos1) # Calcula a soma cumulativa dos passos

    passos2 = np.random.choice([-1, 1], size=N, p=[0.5 , 0.5]) # Gera uma sequência de passos aleatórios
    posições2 = np.cumsum(passos2) # Calcula a soma cumulativa dos passos
    
    
    plt.figure(figsize=(10, 6)) # Define o tamanho da figura
    plt.title("Passeio Aleatório Unidimensional")
    
    plt.subplot(2, 1, 1) # Cria o primeiro subplot
    plt.plot(posições1, label='Passeio Aleatório 1', color='purple') # Plota o passeio aleatório
    plt.xlabel("Número de Passos")
    plt.ylabel("Posição")
    plt.grid(True)
    
    plt.subplot(2, 1, 2) # Cria o segundo subplot
    plt.plot(posições2, label='Passeio Aleatório 2', color='pink') # Plota o passeio aleatório
    plt.xlabel("Número de Passos")
    plt.ylabel("Posição")
    plt.grid(True)
    plt.show() # Exibe o gráfico

    return posições1 # Imprime as posições ao longo do passeio aleatório
    

#teste 
Passeio_aleatorio(1000) # Chama a função com 1000 passos
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Passeio_aleatorio(N=1000, n_simulacoes=100):
    """
    Simula passeios aleatórios unidimensionais e plota as trajetórias e médias.
    
    Parâmetros:
    N (int): Número de passos em cada passeio aleatório.
    n_simulacoes (int): Número de simulações para calcular as médias.
    """
    # Arrays para armazenar todas as simulações
    todas_posicoes1 = np.zeros((n_simulacoes, N))
    todas_posicoes2 = np.zeros((n_simulacoes, N))
    
    # Simula vários passeios aleatórios
    for i in range(n_simulacoes):
        # Passeio com viés (p=0.4/0.6)
        passos1 = np.random.choice([-1, 1], size=N, p=[0.6, 0.4])
        todas_posicoes1[i] = np.cumsum(passos1)
        
        # Passeio sem viés (p=0.5/0.5)
        passos2 = np.random.choice([-1, 1], size=N, p=[0.5, 0.5])
        todas_posicoes2[i] = np.cumsum(passos2)
    
    # Calcula médias
    medias1 = np.mean(todas_posicoes1, axis=0)
    medias2 = np.mean(todas_posicoes2, axis=0)
    
    # Cria figura para os passeios individuais e médias
    plt.figure(figsize=(12, 8))
    
    # Primeiro subplot
    plt.subplot(2, 2, 1)
    for i in range(min(10, n_simulacoes)):  # Plota apenas 10 trajetórias para clareza
        plt.plot(todas_posicoes1[i], alpha=0.5)
    plt.title("10 Trajetórias (p=0.4/0.6)")
    plt.xlabel("Passos")
    plt.ylabel("Posição")
    plt.grid(True)
    
    # Segundo subplot
    plt.subplot(2, 2, 2)
    plt.plot(medias1, color='purple', linewidth=2)
    plt.title(f"Média de {n_simulacoes} Trajetórias (p=0.4/0.6)")
    plt.xlabel("Passos")
    plt.ylabel("Posição Média")
    plt.grid(True)
    
    # Terceiro subplot
    plt.subplot(2, 2, 3)
    for i in range(min(10, n_simulacoes)):
        plt.plot(todas_posicoes2[i], alpha=0.5)
    plt.title("10 Trajetórias (p=0.5/0.5)")
    plt.xlabel("Passos")
    plt.ylabel("Posição")
    plt.grid(True)
    
    # Quarto subplot
    plt.subplot(2, 2, 4)
    plt.plot(medias2, color='blue', linewidth=2)
    plt.title(f"Média de {n_simulacoes} Trajetórias (p=0.5/0.5)")
    plt.xlabel("Passos")
    plt.ylabel("Posição Média")
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    return medias1, medias2

# Teste da função
medias1, medias2 = Passeio_aleatorio(N=1000, n_simulacoes=100)
```

Atividade Intermediária

```python
import numpy as np
import matplotlib.pyplot as plt

def Passeio_aleatorio_media(N, n_simulacoes=100):
    """
    Simula múltiplos passeios aleatórios unidimensionais e plota as médias.
    
    Parâmetros:
    N (int): Número de passos em cada passeio aleatório.
    n_simulacoes (int): Número de simulações para calcular as médias.
    
    Retorna:
    tuple: (medias1, medias2) - Arrays com as médias das posições ao longo do tempo
    """
    # Arrays para armazenar as médias ao longo do tempo
    medias1 = np.zeros(N)
    medias2 = np.zeros(N)
    
    # Arrays para armazenar os desvios padrão
    desvios1 = np.zeros(N)
    desvios2 = np.zeros(N)
    
    for _ in range(n_simulacoes):
        # Simula passeios aleatórios com diferentes probabilidades
        passos1 = np.random.choice([-1, 1], size=N, p=[0.6, 0.4])
        posicoes1 = np.cumsum(passos1)
        
        passos2 = np.random.choice([-1, 1], size=N, p=[0.5, 0.5])
        posicoes2 = np.cumsum(passos2)
        
        # Acumula para cálculo das médias
        medias1 += posicoes1
        medias2 += posicoes2
        
        # Acumula quadrados para cálculo da variância
        desvios1 += posicoes1**2
        desvios2 += posicoes2**2
    
    # Calcula médias e desvios
    medias1 /= n_simulacoes
    medias2 /= n_simulacoes
    
    desvios1 = np.sqrt(desvios1/n_simulacoes - medias1**2)
    desvios2 = np.sqrt(desvios2/n_simulacoes - medias2**2)
    
    # Cria figura para os passeios individuais
    plt.figure(figsize=(12, 8))
    
    # Passeios individuais (apenas 5 exemplos)
    plt.subplot(2, 1, 1)
    for _ in range(5):
        passos = np.random.choice([-1, 1], size=N, p=[0.6, 0.4])
        plt.plot(np.cumsum(passos), alpha=0.5)
    plt.title("Exemplos de Passeios Aleatórios (p=0.4/0.6)")
    plt.xlabel("Número de Passos")
    plt.ylabel("Posição")
    plt.grid(True)
    
    plt.subplot(2, 1, 2)
    for _ in range(5):
        passos = np.random.choice([-1, 1], size=N, p=[0.5, 0.5])
        plt.plot(np.cumsum(passos), alpha=0.5)
    plt.title("Exemplos de Passeios Aleatórios (p=0.5/0.5)")
    plt.xlabel("Número de Passos")
    plt.ylabel("Posição")
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Cria figura para as médias
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(medias1, label='Média', color='blue')
    plt.fill_between(range(N), medias1 - desvios1, medias1 + desvios1, 
                    color='blue', alpha=0.2, label='Desvio Padrão')
    plt.title("Média do Passeio (p=0.4/0.6)")
    plt.xlabel("Número de Passos")
    plt.ylabel("Posição Média")
    plt.legend()
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.plot(medias2, label='Média', color='purple')
    plt.fill_between(range(N), medias2 - desvios2, medias2 + desvios2, 
                    color='green', alpha=0.2, label='Desvio Padrão')
    plt.title("Média do Passeio (p=0.5/0.5)")
    plt.xlabel("Número de Passos")
    plt.ylabel("Posição Média")
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    return medias1, medias2

# Teste com 1000 passos e 100 simulações
medias1, medias2 = Passeio_aleatorio_media(1000, 100)

# Mostra os valores finais
print(f"Média final (p=0.4/0.6): {medias1[-1]:.2f}")
print(f"Média final (p=0.5/0.5): {medias2[-1]:.2f}")
```
